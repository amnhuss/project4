```python
import pandas as pd
import plotly.graph_objects as go
```

```python
# Load file
nonimmsuptable_country = "2022_xxxx_plcy_nonimmigrants_fy2021_supplemental_tables/fy2021_nonimmsuptable1d.xlsx"
nonimmsuptable_state = "2022_xxxx_plcy_nonimmigrants_fy2021_supplemental_tables/fy2021_nonimmsuptable3d.xlsx"


# Read excel
nonimmsuptable_country = pd.read_excel(nonimmsuptable_country, header=None)
nonimmsuptable_state = pd.read_excel(nonimmsuptable_state, header=None)

# Remove first 3 rows that is irralevent and set header and index
nonimmsuptable_country = nonimmsuptable_country.drop(nonimmsuptable_country.index[0:3])
nonimmsuptable_country = nonimmsuptable_country.reset_index(drop=True)
nonimmsuptable_country = nonimmsuptable_country.set_axis(nonimmsuptable_country.iloc[0],axis = "columns")
nonimmsuptable_country = nonimmsuptable_country[2:]
# nonimmsuptable_country = nonimmsuptable_country.set_index(nonimmsuptable_country.columns[0])
nonimmsuptable_country = nonimmsuptable_country.drop("Total", axis=1)

nonimmsuptable_state = nonimmsuptable_state.drop(nonimmsuptable_state.index[0:3])
nonimmsuptable_state = nonimmsuptable_state.reset_index(drop=True)
nonimmsuptable_state = nonimmsuptable_state.set_axis(nonimmsuptable_state.iloc[0],axis = "columns")
nonimmsuptable_state = nonimmsuptable_state[2:]
# nonimmsuptable_state = nonimmsuptable_state.set_index(nonimmsuptable_state.columns[0])
nonimmsuptable_state = nonimmsuptable_state.drop("Total", axis=1)
```
